# Imports

```python
import os
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

from shapely.geometry import Point
```

# Load Datasets

### Admin Boundaries

```python
ph = gpd.read_file('gadm36_PHL_shp/gadm36_PHL_1.shp')
```

```python
ph.head()
```

### Health Sites

```python
hs = gpd.read_file('philippines.geojson', driver='GeoJSON')
hs.head()
```

## Inspection

### Projection

Upon checking, both datasets follow the same projection

```python
ph.crs
```

```python
hs.crs
```

### Rename Columns

For convenience, rename the province name column

```python
ph.head()
```

```python
ph.rename(columns={"NAME_1": "Province"}, inplace=True)
```

```python
ph.head(2)
```

# DataViz

### Nationwide

Plot all data points in `hs` dataset into the `ph` map plot

```python
fig = plt.figure(figsize=(25, 20))
ax = fig.add_subplot(111)
ph.plot(ax=ax, color='white', edgecolor='dimgray')
hs.plot(ax=ax, column="amenity", alpha=0.35)
ax.set_title('Health Amenities all over the Philippines', fontsize=15)
ax.axis('off');
```

### Isolating Selected Provinces

My family and my father-side relatives live in Cavite. My mother-side relatives live in Laguna

# Cavite

```python
cavite = ph[ph['Province'] == 'Cavite']
```

```python
ax = cavite.plot(figsize=(5,5), color="w", edgecolor="dimgray")
```

#### Find the amenities within Cavite

```python
cavite_hs = gpd.sjoin(hs, cavite, op='within')
cavite_hs.head()
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111)
cavite.plot(ax=ax, color='white', edgecolor='dimgray')
cavite_hs.plot(ax=ax, column="amenity", alpha=0.35, legend=True)
ax.set_title('Amenities in Cavite', fontsize=15)
ax.axis('off');
```

### Laguna

```python
laguna = ph[ph['Province'] == 'Laguna']
```

```python
ax = laguna.plot(figsize=(5,5), color="w", edgecolor="dimgray")
```

### Find amenities within Quezon province

```python
laguna_hs = gpd.sjoin(hs, laguna, op='within')
laguna_hs.head()
```

```python
fig = plt.figure(figsize=(20, 15))
ax = fig.add_subplot(111)
laguna.plot(ax=ax, color='white', edgecolor='dimgray')
laguna_hs.plot(ax=ax, column="amenity", alpha=0.35, legend=True)
ax.set_title('Amenities in Laguna', fontsize=15)
ax.axis('off');
```

# Breaking down into amenities

```python
# excluding Quezon because it seems to have a far outlying area that 
# makes the plot zoom out too far just to display the whole area
# Since these provinces are next to each other, I'll just include Cavite and Laguna
# feel free to use the other list to see Batangas and Rizal (but not the amenities)
# CALABARZON = ["Cavite", "Laguna", "Batangas", "Rizal"]
CALABARZON = ["Cavite", "Laguna"]
```

```python
reg4a = ph[ph['Province'].isin(CALABARZON)] 
```

### All Amenities

```python
fsize = (25, 20)
fig = plt.figure(figsize=fsize)
ax = fig.add_subplot(111)
reg4a.plot(ax=ax, color='black', edgecolor='dimgray')
cavite.plot(ax=ax, color='white', edgecolor='dimgray', legend=True)
laguna.plot(ax=ax, color='lightgray', edgecolor='dimgray')
cavite_hs.plot(ax=ax, column="amenity", alpha=0.35, legend=True)
laguna_hs.plot(ax=ax, column="amenity", alpha=0.35)
ax.set_title('Health Amenities in Cavite (White Area) and Laguna (Gray Area)', fontsize=25)
#ax.axis('off');
```

### Dentists

```python
clr = "blue"
fig = plt.figure(figsize=fsize)
ax = fig.add_subplot(111)
reg4a.plot(ax=ax, color='black', edgecolor='dimgray')
cavite.plot(ax=ax, color='white', edgecolor='dimgray', legend=True)
laguna.plot(ax=ax, color='lightgray', edgecolor='dimgray')
cavite_hs[cavite_hs['amenity'] == 'dentist'].plot(ax=ax, color=clr, alpha=0.35)
laguna_hs[laguna_hs['amenity'] == 'dentist'].plot(ax=ax, color=clr, alpha=0.35)
ax.set_title('Dentist in Cavite (White Area) and Laguna (Gray Area)', fontsize=25)
#ax.axis('off');
```
